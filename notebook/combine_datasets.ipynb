```python
splits = ['train', 'valid', 'test']
subfolders = ['images', 'labels']

for split in splits:
    print(f"\n--- {split.upper()} ---")
    for sub in subfolders:
        src1 = os.path.join(ds1_root, split, sub)
        src2 = os.path.join(ds2_root, split, sub)
        dst  = os.path.join(merged_root, split, sub)
        print('dst', dst)

        if os.path.isdir(src1):
            print(f'Số file trong {src1}: {len(os.listdir(src1))}')
        if os.path.isdir(src2):
            print(f'Số file trong {src2}: {len(os.listdir(src2))}')

        os.makedirs(dst, exist_ok=True)

        copied_files = set()

        # Copy tất cả file từ dataset1
        if os.path.isdir(src1):
            for fname in os.listdir(src1):
                target_path = os.path.join(dst, fname)
                shutil.copy2(os.path.join(src1, fname), target_path)
                copied_files.add(fname) # Thêm vào set

        # Copy tất cả file từ dataset2
        if os.path.isdir(src2):
            num_overwritten = 0
            num_new = 0
            for fname in os.listdir(src2):
                target_path = os.path.join(dst, fname)

                if fname in copied_files:
                    num_overwritten += 1 # Đếm số file bị ghi đè
                    # Optionally, you can rename here instead of overwriting
                    # new_fname = f"ds2_{fname}"
                    # shutil.copy2(os.path.join(src2, fname), os.path.join(dst, new_fname))
                else:
                    num_new += 1 # Đếm số file mới

                shutil.copy2(os.path.join(src2, fname), target_path) # Vẫn copy, sẽ ghi đè nếu trùng

            print(f"  --> {split} {sub}: Overwritten from DS2: {num_overwritten}, New from DS2: {num_new}")

        print(f"Tổng số file sau khi gộp trong {split} {sub}: {len(os.listdir(dst))}")

print("\nĐã gộp xong vào:", merged_root)

label_train_merged_path = os.path.join(merged_root, 'train', 'labels')
label_test_merged_path = os.path.join(merged_root, 'test', 'labels')
label_valid_merged_path = os.path.join(merged_root, 'valid', 'labels')

images_train_merged_path = os.path.join(merged_root, 'train', 'images')
images_test_merged_path = os.path.join(merged_root, 'test', 'images')
images_valid_merged_path = os.path.join(merged_root, 'valid', 'images')

total_labels = 0
if os.path.isdir(label_train_merged_path): total_labels += len(os.listdir(label_train_merged_path))
if os.path.isdir(label_test_merged_path): total_labels += len(os.listdir(label_test_merged_path))
if os.path.isdir(label_valid_merged_path): total_labels += len(os.listdir(label_valid_merged_path))
print(f"\nTổng số file labels đã gộp: {total_labels}")

total_images = 0
if os.path.isdir(images_train_merged_path): total_images += len(os.listdir(images_train_merged_path))
if os.path.isdir(images_test_merged_path): total_images += len(os.listdir(images_test_merged_path))
if os.path.isdir(images_valid_merged_path): total_images += len(os.listdir(images_valid_merged_path))
print(f"Tổng số file images đã gộp: {total_images}")
```
